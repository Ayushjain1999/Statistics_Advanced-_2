```python
'''
Question 1: What is hypothesis testing in statistics?
Answer:     Hypothesis testing is a statistical method used to make inferences about a population based on sample data. 
            It involves formulating two competing hypotheses: the null hypothesis (H₀) and the alternative hypothesis (H₁). 
            A test statistic is calculated from the sample, and based on its value and a chosen significance level, 
            we either reject H₀ or fail to reject H₀.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
Answer:  a. Null Hypothesis (H₀): A statement that assumes no effect, no difference, or no relationship exists. It is the default assumption.
         b. Alternative Hypothesis (H₁): A statement that contradicts H₀, suggesting that an effect, difference, or relationship does exist.
            Example: Testing whether the average weight of apples = 100g.
            H₀: μ = 100
            H₁: μ ≠ 100

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
Answer:     The significance level (α) is the probability threshold for rejecting H₀. Commonly set at 0.05 or 0.01, 
            it represents the maximum probability of committing a Type I error (rejecting H₀ when it is true). 
            If the p-value ≤ α, we reject H₀; otherwise, we fail to reject it.

Question 4: What are Type I and Type II errors? Give examples of each.
Answer:  a. Type I Error (False Positive): Rejecting H₀ when it is true. Example: Declaring a medicine effective when it is not.
         b. Type II Error (False Negative): Failing to reject H₀ when H₁ is true. Example: Concluding a medicine has no effect when it actually does.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
Answer:  a. Z-test: Used when population variance (σ²) is known and sample size is large (n > 30).
         b. T-test: Used when population variance is unknown and sample size is small (n < 30).
            Both test means, but T-test adjusts for small-sample uncertainty.
'''
```

```python
#Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
#Answer:

import numpy as np
import matplotlib.pyplot as plt
n, p = 10, 0.5
data = np.random.binomial(n, p, 1000)
plt.hist(data, bins=range(n+2), edgecolor='black', alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
#Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
#sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, #50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
#Answer:

import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
mu_0 = 50
mean = np.mean(sample_data)
std = np.std(sample_data, ddof=1)
n = len(sample_data)
z_stat = (mean - mu_0) / (std / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))
print("Mean:", mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
```

```python
#Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
#Answer:

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
data = np.random.normal(loc=50, scale=5, size=1000)
mean = np.mean(data)
se = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=se)
print("Mean:", mean)
print("95% CI:", ci)
plt.hist(data, bins=30, alpha=0.7, edgecolor="black")
plt.axvline(ci[0], color='red', linestyle='--', label="Lower 95% CI")
plt.axvline(ci[1], color='red', linestyle='--', label="Upper 95% CI")
plt.axvline(mean, color='blue', linestyle='-', label="Mean")
plt.legend()
plt.title("Normal Distribution with 95% CI")
plt.show()
```

```python
#Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. 
#Explain what the Z-scores represent in terms of standard deviations from the mean.
#Answer:

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
data = np.random.normal(100, 15, 1000)
z_scores = zscore(data)
plt.hist(z_scores, bins=30, edgecolor="black", alpha=0.7)
plt.title("Standardized Data (Z-scores)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
print("Z-scores represent how many standard deviations each point is from the mean.")
```
